# Überprüfung der bereinigten Daten

```python
import duckdb as dd
```

```python
con = dd.connect(r"data.db")
```

```python
# Beispiel Aufruf über die Tabellen Customer und City
con.execute(
    """
    SELECT 
      customer_id, 
      first_name, 
      last_name, 
      city.city as city 
    FROM 
      Customer 
    JOIN City ON Customer.city_id = City.city_id 
    WHERE First_Name = 'Alex'
    LIMIT 10"""
).df()
```

```python
con.execute("describe select * from Usage;").df()
```

```python
con.execute(
    """select * from Usage where "Usage Time" < TIMESTAMP '1900-01-01 00:00:00'"""
).df()
```

```python
con.execute(
    """
    SELECT count(*) FROM Usage;
    """
).df()
```

```python
con.execute(
    """
    SELECT * FROM Usage LIMIT 10;    
    """
).df()
```

```python
con.execute(
    """
    SELECT * FROM Customer LIMIT 10;    
    """
).df()
```

```python
con.execute(
    """
    SELECT * FROM Service LIMIT 10;    
    """
).df()
```

```python
con.execute(
"""
select distinct service_id from Usage where service_id not in (select service_id from Service) 
"""
).df()
```

```python
con.execute("select distinct city_id from Customer where city_id not in (select city_id from City)").df()
```

```python
con.execute("select * from Customer where city_id is Null").df()
```

```python
con.execute("""SELECT customer_id, COUNT(*) 
FROM Customer 
GROUP BY customer_id 
HAVING COUNT(*) > 1;""").df()
```
