```python
import duckdb as dd
```

```python
con = dd.connect("data.db")
```

```python
con.execute(
    """
WITH duplicates AS (
  SELECT 
    service_name,
    count(*) as counter
  FROM 
    Service
  GROUP BY service_name
  HAVING counter > 1
)
SELECT s.costs_per_month, s.service_id, d.counter FROM SERVICE s
JOIN duplicates d ON d.service_name = s.service_name  
"""
).df()
```

```python
con.execute(
"""
WITH duplicates AS (
  SELECT
  ROW_NUMBER() OVER (PARTITION BY Service_name) as counter,
  service_name
  FROM Service
) SELECT * FROM duplicates WHERE counter > 1
"""
).df()
```
